### Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function `y=(x+3)²` starting from the point `x=2`.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(x):
    return (x + 3) ** 2

def grad(x):
    return 2 * (x + 3)
```

```python
x = 2
lr = 0.1
max_iters = 50
x_history = [x]
y_history = [fn(x)]
```

```python
for i in range(max_iters):
    x = x - lr * grad(x)

    x_history.append(x)
    y_history.append(fn(x))
    
print(x)
print(fn(x))
```

```python
x_range = np.linspace(-9, 3, 200)
y_range = fn(x_range)

# plt.figure(figsize= (200, 200))

plt.plot(x_range, y_range)
plt.scatter(x_history, y_history, color= 'r')
plt.plot(x_history, y_history, linestyle='-.', color='y')
plt.xlabel('x')
plt.ylabel('y = (x + 3)²')
```

```python
plt.plot(range(len(y_history)), y_history)
```
